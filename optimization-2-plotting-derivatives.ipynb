# Finding And Plotting Derivatives Of A Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting a function:

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_func(func):
    # Generate 1000 equally spaced x values between -10 and 10
    x_values = np.linspace(-10, 10, 1000)
    
    # Evaluate the input function at each x value to generate corresponding y values
    y_values = [func(x) for x in x_values]
    
    # Plot the x and y values using Matplotlib
    plt.plot(x_values, y_values)
```

```python
plot_func(lambda x: x ** 2)
```

### Passing the function:

```python
plot_func(lambda x: x ** 2 + 50)
```

### Finding Derivative of the Function at k:

```python
def derivate(func, k):
    # Define the delta value for finding the derivative
    delta_k = 0.000001
    
    # Calculate the derivative of the input function at k
    derivative = (func(k + delta_k) - func(k)) / delta_k
    
    # Return the calculated derivative
    return derivative
```

```python
# Calculate the derivative of x^2 at x = 5 using the derivate() function
derivate(lambda x: x ** 2, 5)

# The derivative of f(x) = x^2 is f'(x) = 2x, which means that the derivative of f(x) at x=5 is f'(5) = 2(5) = 10.
```

```python
derivate(lambda x: x ** 2, 10)
```

```python
derivate(lambda x: x ** 2, 0) 

# At x=0, the curve y=x^2 has a horizontal tangent line (i.e., a slope of 0), 
# which means that the derivative at x=0 is also 0.
```

```python
derivate(lambda x: x ** 2, -7.5)

# It means that with change in x, change in y is negative.
```

### Plotting Derivative:

```python
def plot_derivate(func):
    # Generate a list of x-values to evaluate the derivative at
    x_values = np.linspace(-10, 10, 1000)
    
    # Calculate the derivative of the input function for each x-value in x_values
    deriv_values = [derivate(func, x) for x in x_values]
    
    # Plot the derivative values against the corresponding x-values
    plt.plot(x_values, deriv_values)
```

```python
plot_derivate(lambda x: x ** 2)
```

```python
plot_func(lambda x: x ** 3)
```

```python
plot_derivate(lambda x: x ** 3)
```
